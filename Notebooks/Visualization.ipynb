```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
Data = pd.read_csv('/home/omnix/Desktop/E-Commerce/E-commerce_EDA/Data/Cleaned.csv')
Data
```

```python
sns.histplot(Data['Revenue'],bins=200,kde=True)
plt.title("Distribution of Revenue ")
```

```python
sns.histplot(Data['discount_amount'],bins=100,kde=True)
plt.title("Distribution of Discount Among Products ")
```

```python
sns.histplot(Data['subtotal'],kde=True)
plt.title('Distribution of Subtotal ')
```

```python
sns.histplot(Data['shipping_cost'],kde=True)
plt.title('Distribution of Shiping Cost')
```

```python
sns.histplot(Data['net_amount'],kde=True)
plt.title("Distribution of the Net Amount ")
```

```python
sns.histplot(Data['unit_price'],kde=True)
plt.title('Distribution of the Unit Price ')
```

```python
customer_segment_Revenue = Data.groupby('customer_segment')['Revenue'].sum().sort_values(ascending=False)
customer_segment_Revenue.plot(kind='bar')
plt.title("Revenue According to the Customer segment ")
plt.grid(True)
# Conlcusion : Customers With Standard Segment Generate the Highest Revenue Followed By Silver Segment 
```

```python
# Product_name 
product_name_Revenue = Data.groupby('product_name')['Revenue'].sum().sort_values(ascending=False)
product_name_Revenue.plot(kind='barh')
plt.title("Revenue Acording Different Products")
plt.grid(True)

# Conclusion : Laptops Generates the Highest Revenue Followed By the Tablet 
```

```python
Data
```

```python
Shipping_cost_City = Data.groupby ('shipping_city')['shipping_cost'].sum().sort_values(ascending=False)
plt.figure(figsize=(20,12))
Shipping_cost_City.plot(kind='bar')
plt.title("Shiping Cost Acording to Location")
# Conclusion : Shiiping Costs Are the Highest for Locations in Chennai Followed By Pune 
```

```python
Payement_Methods_Revenue = Data.groupby('payment_method')['Revenue'].sum().sort_values(ascending=True)
Payement_Methods_Revenue.plot(kind="barh")
plt.title('Revnue according to the Payment getway ')

# Conclusion : Payment Gateway (Credit Card ) generates Highest Revenue Followed By Debit Card  
```

```python
Year_Wise_Sales = Data.groupby('Year')['Revenue'].sum().sort_values(ascending=False)
Year_Wise_Sales.plot(kind='barh')
plt.title('Sales For Last Two Years ')
# Conclusion  : Sales increased By nearly 7-8% From the year 2023
```

```python
# What Affect Sales 
Sales_data = Data[['quantity','unit_price','discount_percent','shipping_cost','Year','Month','Day',\
                   'rating','subtotal',	'discount_amount',	'net_amount'	,'Revenue']]
plt.figure(figsize=(20,12))
sns.heatmap(Sales_data.corr(),annot=True)




# CONCLUSION: Unit Price, Subtotal, Discount Amount, and Net Amount are the STRONGEST predictors of Revenue.
```
